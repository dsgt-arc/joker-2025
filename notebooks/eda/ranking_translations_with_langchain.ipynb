```python
from langchain_anthropic import ChatAnthropic

max_tokens = 1024
translation_instructions = """
Step 1: Identify the key words in the pun that contribute to its humor or wordplay. Focus on words with double meanings, homophones, homonyms, or 
phonetic similarities that create ambiguity or a clever twist.
Step 2: Provide a natural language explanation of the humor or wordplay. Describe how the identified words interact to create a pun, explaining any cultural, phonetic, or syntactic nuances that make it humorous.
Step 3: Generate five different French translations of the pun. Each translation should use a different humor mechanism.
Step 4: Provide english translations of the French puns
Step 4: Rank the translations from 1 to 5 based on how well they align with french humor patterns
"""

pun ="Input: How do spacemen add more protein to their diet ? They make it meteor ."

input = translation_instructions + pun


model = 'claude-3-5-sonnet-20240620'
api_key = ""
llm = ChatAnthropic(model=model, api_key=api_key, max_tokens=max_tokens, temperature=1)
translation_response = llm.invoke(input)
print(model, '\n', translation_response.content, '\n')
```

```python
llm = ChatAnthropic(model=model, api_key=api_key, max_tokens=max_tokens, temperature=0.5)
translation_response = llm.invoke(input)
print(model, '\n', translation_response.content, '\n')
```

```python
llm = ChatAnthropic(model=model, api_key=api_key, max_tokens=max_tokens, temperature=0)
translation_response = llm.invoke(input)
print(model, '\n', translation_response.content, '\n')
```
